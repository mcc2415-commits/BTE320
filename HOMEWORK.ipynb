```python
def getValues(n):
    values = []
    i = 0
    while i < n:
        x = float(input("Enter a number: "))
        values.append(x)
        i = i + 1
    return values

def calcAverage(alist):
    total = 0
    count = 0

    i = 2
    while i < len(alist):
        total = total + alist[i]
        count = count + 1
        i = i + 1
    if count == 0:
        return 0
    return total / count

n = int(input("How many numbers: "))
nums = getValues(n)
print("Average is:", calcAverage(nums))
```

```python
def getParkingHours():
  hours = float(input("Enter # of hours parked: "))
  return hours

def calcParkingFee(hours):
  cost = hours * 2.5
  if cost > 20:
    cost = 20
    return cost
  elif cost < 6:
    cost = 6
    return cost
  else:
    return cost

hours = getParkingHours()
cost = calcParkingFee(hours)

print(f"Your Fee is", cost)
```

```python
def calcTuition(credits):
  if credits >= 12:
    price = 20000
    return price
  else:
    price = (1700 * credits) + 1200
    return price

credits = int(input("How many credits: "))
price = calcTuition(credits)
print("Your tuition is", price)
```
