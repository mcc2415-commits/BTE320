```python
import random

number = random.randint(1, 2)
number2 = random.randint(3, 4)

if number == 1:
  number = "URGENT"
else:
 number = "Hello"

if number2 == 3:
  number2 = "www.phis.org"
else:
 number2 = "www.worldpeace.org"

email = "{number}!! Click on the following: {number2} to ...".format(number=number, number2=number2)

is_urgent = "URGENT"
has_www = "www.phis.org"

if is_urgent in email and has_www in email:
  print("Phishing Attack!")
if is_urgent in email and not has_www in email:
  print("Spam")
if not is_urgent in email and has_www in email:
  print("Dangerous Link")
if not is_urgent in email and not has_www in email:
  print("Safe")

print(email)
```
