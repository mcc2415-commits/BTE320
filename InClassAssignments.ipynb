```python
sales = float(input('What are the Sales: '))
TR = float(input('What is the Tax Rate: ')) /100
COGS = sales * 0.4
GP = sales - COGS
taxes = GP * TR
net = GP - taxes
print(f'Sales: {sales}\nCost of Goods Sold: ${COGS}\nGross Profit: ${GP}\nTaxes: {taxes}\nNet Income: {net}')
```

```python
Num1 = float(input("What is your First Number: "))
Num2 = float(input("What is your Second Number: "))
Sign =  input("What Operation (*,/,-,+,**): ")
if Sign == "*":
  answer = Num1 * Num2
  print(f"{Num1} multiplied by {Num2} = {answer}")
elif Sign == "/":
  answer = Num1 / Num2
  print(f"{Num1} divided by {Num2} = {answer}")
elif Sign == "-":
   answer = Num1 - Num2
   print(f"{Num1} subtracted by {Num2} = {answer}")
elif Sign == "+":
  answer = Num1 + Num2
  print(f"{Num1} added to {Num2} = {answer}")
elif Sign == "**":
  answer = Num1 ** Num2
  print(f"{Num1} to the power of {Num2} = {answer}")
else:
  print("Use one of the listed operations")
```

```python
def check(email):
  assert type(email) == str, "input a string"
  i = len(email)
  c = 0
  d = 0
  for i in range(0,i,1):
    if email[i] == "@":
      c += 1
      if c > 1:
        print("False, Invalid Email")
        return
    if email[i] == ".":
      d += 1
  if d >= 1 and c == 1:
    print("True, Valid Email")
  else:
    print("False, Invalid Email")


email = input("What is you email: ")
check(email)
```
